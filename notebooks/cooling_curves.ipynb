```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./mpl_style")
import pylbo
from pylbo.utilities.debug import get_debug_coolingcurves
```

```python
def add_coolcurve_to_figure(ax, file, linestyle=".", color="black", dotcolor="black"):
  coolcurve = get_debug_coolingcurves(file)
  ax.loglog(
    coolcurve["T_curve"],
    coolcurve["L_curve"],
    linestyle,
    color=color,
    alpha=0.8,
    lw=3,
  )
  ax.loglog(
    coolcurve["T_table"],
    coolcurve["L_table"],
    ".",
    color=dotcolor,
    alpha=1,
    markersize=7,
  )
```

```python
  
fig, ax = plt.subplots(1, figsize=(10, 6))

add_coolcurve_to_figure(
  ax, 
  file="files/coolingcurve_spexdm", 
  linestyle="-", 
  color="lightsalmon", 
  dotcolor="red",
)
add_coolcurve_to_figure(
  ax, 
  file="files/coolingcurve_jccorona", 
  linestyle="-", 
  color="deepskyblue", 
  dotcolor="navy",
)

ax.set_xlabel("temperature [K]")
ax.set_ylabel(r"$\Lambda(T)$ [erg s$^{-1}$ cm$^3$]")
ax.set_ylim(1e-28, 1e-20)
ax.set_xlim(10, 1e8)
ax.legend(
  [
    "Spex DM (interpolated)", 
    "Spex DM (table)", 
    "JC Corona (interpolated)",
    "JC Corona (table)",
  ], 
  loc="lower right",
)
fig.tight_layout()
fig.savefig("../03-thermal_instability/figures/coolingcurves.png")
```
