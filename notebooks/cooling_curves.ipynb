```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./mpl_style")
import pylbo
from pylbo.utilities.debug import get_debug_coolingcurves
```

```python
def add_coolcurve_to_figure(ax, file, linestyle=".", color="black"):
  coolcurve = get_debug_coolingcurves(file)
  ax.loglog(
    coolcurve["T_table"],
    coolcurve["L_table"],
    linestyle,
    color=color,
    alpha=0.8,
    lw=2,
  )
```

```python
  
fig, ax = plt.subplots(1, figsize=(10, 6))

add_coolcurve_to_figure(
  ax, file="files/coolingcurve_spexdm", linestyle="-", color="red"
)
add_coolcurve_to_figure(
  ax, file="files/coolingcurve_jccorona", linestyle="-", color="blue"
)

ax.set_xlabel("temperature [K]")
ax.set_ylabel(r"$\Lambda(T)$ [erg s$^{-1}$ cm$^3$]")
ax.set_ylim(1e-28, 1e-20)
ax.set_xlim(10, 1e8)
ax.legend(["Spex DM", "JC Corona"], loc="lower right")
fig.tight_layout()
fig.savefig("../03-thermal_instability/figures/coolingcurves.png")
```
