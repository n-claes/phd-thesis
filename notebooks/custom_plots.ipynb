```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./mpl_style")

slowlabel = r"$\omega_s$"
alfvenlabel = r"$\omega_A$"
fastlabel = r"$\omega_f$"
thermallabel = r"$\omega_{th}$"

xlabel = "Re($\omega$)"
ylabel = "Im($\omega$)"
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))

ax.axhline(y=0, linestyle="dotted", color="grey", alpha=0.8)
ax.axvline(x=0, linestyle="dotted", color="grey", alpha=0.8)

ax.scatter([-4, 4], [0, 0], color="blue", label=fastlabel)
ax.scatter([-2, 2], [0, 0], color="black", label=alfvenlabel)
ax.scatter([-1, 1], [0, 0], color="red", label=slowlabel)
ax.scatter(0, 0, color="green", label=thermallabel)

ax.set_yticks([])
ax.set_xticks([])
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.legend(
  bbox_to_anchor=(0.0, 1.01, 1, 0.102),
  loc="lower left",
  ncol=4,
  mode="expand",
  borderaxespad=0,
)

fig.savefig(
  "../02-MHD_spectroscopy/figures/spectrum_adiabatic.png", 
  dpi=400, 
  bbox_inches="tight"
)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))

ax.axhline(y=0, linestyle="dotted", color="grey", alpha=0.8)
ax.axvline(x=0, linestyle="dotted", color="grey", alpha=0.8)

arrowprops = dict(arrowstyle="->", color="grey", lw=2, alpha=0.8)

# fast
ax.scatter([-4, 4], [0, 0], color="blue", facecolor="none")
ax.scatter([-4, 4], [-0.5, -0.5], color="blue", label=fastlabel)
ax.annotate(text="", xy=(-4, -0.4), xytext=(-4, -0.1), arrowprops=arrowprops)
ax.annotate(text="", xy=(4, -0.4), xytext=(4, -0.1), arrowprops=arrowprops)
# alfven
ax.scatter([-2, 2], [0, 0], color="black", label=alfvenlabel)
# slow
ax.scatter([-1, 1], [0, 0], color="red", facecolor="none")
ax.scatter([-1, 1], [-1, -1], color="red", label=slowlabel)
ax.annotate(text="", xy=(-1, -0.9), xytext=(-1, -0.1), arrowprops=arrowprops)
ax.annotate(text="", xy=(1, -0.9), xytext=(1, -0.1), arrowprops=arrowprops)
# thermal
ax.scatter(0, 0, color="green", facecolor="none")
ax.scatter(0, 1, color="green", label=thermallabel)
ax.annotate(text="", xy=(0, 0.9), xytext=(0, 0.1), arrowprops=arrowprops)

ax.set_yticks([])
ax.set_xticks([])
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.legend(
  bbox_to_anchor=(0.0, 1.01, 1, 0.102),
  loc="lower left",
  ncol=4,
  mode="expand",
  borderaxespad=0,
)

fig.savefig(
  "../02-MHD_spectroscopy/figures/spectrum_nonadiabatic.png", 
  dpi=400, 
  bbox_inches="tight"
)

plt.show()
```
